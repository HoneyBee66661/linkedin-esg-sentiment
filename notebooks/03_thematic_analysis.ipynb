```python
# Template Codebook
codebook = {
    "metadata": {
        "researcher": "Joni Subakti",
        "framework": "Reflexive Thematic Analysis (Braun & Clarke, 2006)",
        "data_source": "LinkedIn posts from extreme sentiment (top/bottom 10%)",
    },
    "initial_codes": [
        {
            "code": "RENEWABLE_ENERGY",
            "description": "Postingan tentang inisiatif energi terbarukan (HVO, solar, biofuel)",
            "example_quote": "Inovasi Energi Terbarukan dalam Setiap Langkah Kami — pengujian renewable diesel HVO pada truk tambang 100 ton...",
            "esg_pillar": "E"
        },
        {
            "code": "CSR_HEALTH",
            "description": "Program CSR bidang kesehatan masyarakat",
            "example_quote": "Perahu Desa Siaga membantu 208 KK mendapatkan akses layanan kesehatan...",
            "esg_pillar": "S"
        },
        {
            "code": "AWARD_ESG",
            "description": "Penghargaan terkait ESG dan tata kelola",
            "example_quote": "Kembali meraih penghargaan The Best Basic Materials Sector dalam CSA Awards 2024...",
            "esg_pillar": "G"
        },
        {
            "code": "COP29_PARTICIPATION",
            "description": "Partisipasi CEO di forum internasional COP29",
            "example_quote": "CEO Dialogue on Climate Change di COP29 Baku...",
            "esg_pillar": "G"
        },
        {
            "code": "WASTE_INNOVATION",
            "description": "Inovasi pengelolaan limbah dan ekonomi sirkular",
            "example_quote": "Inovasi Limbah Plastik Jadi Solusi Ramah Lingkungan — kontainer bibit dari limbah plastik...",
            "esg_pillar": "E"
        },
        {
            "code": "COMMUNITY_EMPOWERMENT",
            "description": "Pemberdayaan ekonomi masyarakat dan UMKM",
            "example_quote": "Memberkenalkan produk olahan UMKM dari Luwu Timur...",
            "esg_pillar": "S"
        },
    ],
    "themes": {
        "Tema 1: Green Initiatives dan Narasi Keberlanjutan": {
            "codes": ["RENEWABLE_ENERGY", "WASTE_INNOVATION"],
            "description": "Postingan yang mengomunikasikan inovasi ramah lingkungan, energi terbarukan, dan ekonomi sirkular",
            "esg_focus": "E"
        },
        "Tema 2: Tanggung Jawab Sosial dan Pemberdayaan Komunitas": {
            "codes": ["CSR_HEALTH", "COMMUNITY_EMPOWERMENT"],
            "description": "Program CSR, pemberdayaan masyarakat, kesehatan, dan pendidikan",
            "esg_focus": "S"
        },
        "Tema 3: Tata Kelola, Transparansi, dan Kepemimpinan": {
            "codes": ["AWARD_ESG", "COP29_PARTICIPATION"],
            "description": "Penghargaan tata kelola, partisipasi forum internasional, dan transparansi",
            "esg_focus": "G"
        }
    },
    "reflectivity_notes": [
        "Sebagian besar data berasal dari akun resmi perusahaan (bukan komentar pengguna independen)",
        "Narasi yang dominan adalah komunikasi satu arah dari korporasi ke publik",
        "Potensi greenwashing perlu diverifikasi melalui triangulasi dengan laporan ESG aktual"
    ]
}

# Save codebook
with open(RESULT_DIR / 'thematic_codes.json', 'w') as f:
    json.dump(codebook, f, indent=2, ensure_ascii=False)

print('Thematic codebook saved to:', RESULT_DIR / 'thematic_codes.json')
print(f'\nInitial codes: {len(codebook["initial_codes"])}')
print(f'Themes: {len(codebook["themes"])}')
```

```python
# Visual: Thematic Map
import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph()

# Add nodes
for code in codebook['initial_codes']:
    G.add_node(code['code'], type='code', pillar=code['esg_pillar'])

for theme_name, theme_data in codebook['themes'].items():
    short_name = theme_name.split(':')[0]  # shorten
    G.add_node(short_name, type='theme')
    for c in theme_data['codes']:
        G.add_edge(c, short_name)

fig, ax = plt.subplots(figsize=(12, 8))
pos = nx.spring_layout(G, k=2, iterations=50)

code_nodes = [n for n, d in G.nodes(data=True) if d.get('type') == 'code']
theme_nodes = [n for n, d in G.nodes(data=True) if d.get('type') == 'theme']

nx.draw_networkx_nodes(G, pos, nodelist=code_nodes, node_color='lightblue', node_size=2000, ax=ax)
nx.draw_networkx_nodes(G, pos, nodelist=theme_nodes, node_color='lightgreen', node_size=3000, ax=ax)
nx.draw_networkx_edges(G, pos, edge_color='gray', arrows=True, ax=ax)
nx.draw_networkx_labels(G, pos, font_size=9, ax=ax)

ax.set_title('Thematic Map: ESG Narratives on LinkedIn', fontsize=14)
ax.axis('off')
plt.tight_layout()
plt.savefig(FIGURE_DIR / '06_thematic_map.png', dpi=300, bbox_inches='tight')
plt.show()
```
